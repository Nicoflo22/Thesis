```python
from tensorflow.keras import layers
from tensorflow.keras.models import Model
import tensorflow as tf

def convolutional_blocks(inputs, filters, kernel_size, dropout=0):
    x = layers.Conv1D(filters=filters, kernel_size=kernel_size, padding='same', activation='relu')(inputs)
    x = layers.BatchNormalization()(x)
    x = layers.Dropout(dropout)(x)
    return x

def transformer_encoder(inputs, head_size, num_heads, ff_dim, dropout=0):
    x = convolutional_blocks(inputs, filters=32, kernel_size=7, dropout=dropout)
    x = convolutional_blocks(x, filters=64, kernel_size=5, dropout=dropout)
    x = convolutional_blocks(x, filters=128, kernel_size=3, dropout=dropout)

    # Duplicate the processed signal
    x1, x2 = tf.split(x, num_or_size_splits=2, axis=-1)

    # Apply attention to the first half of the signal
    x1 = layers.LayerNormalization(epsilon=1e-6)(x1)
    x1 = layers.MultiHeadAttention(
        key_dim=head_size, num_heads=num_heads, dropout=dropout
    )(x1, x1)

    # Apply a different transformation to the second half of the signal
    x2 = layers.LayerNormalization(epsilon=1e-6)(x2)
    x2 = layers.Conv1D(filters=ff_dim, kernel_size=1, activation="relu")(x2)
    x2 = layers.Dropout(dropout)(x2)
    x2 = layers.Conv1D(filters=x1.shape[-1], kernel_size=1)(x2)  # Ensure the output size matches x1

    # Calculate the dot product between the two halves
    x = tf.einsum('...i,...i->...', x1, x2)
    return x


inputs = layers.Input(shape=(2500, 12))
x = transformer_encoder(inputs, head_size=64, num_heads=2, ff_dim=2)
outputs = layers.Dense(5, activation="softmax")(x)

model = tf.keras.Model(inputs=inputs, outputs=outputs)
model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=0.001),
              loss='categorical_crossentropy',
              metrics=['categorical_accuracy'])

# Convert the lists to numpy arrays
X_train = np.array(X_ecg_train)
X_val = np.array(X_ecg_val)
X_test = np.array(X_ecg_test)

# Fit the model
model.fit(X_train, y_train, validation_data=(X_val, y_val),
          epochs=10, batch_size=32)
```
